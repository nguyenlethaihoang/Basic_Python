```python
import matplotlib.pyplot as plt 
plt.style.use('default')
years = [2013, 2014, 2015, 2016, 2017, 2018, 2019, 2020]
#profit of CompanyA
companyA = [41596, 46000, 48510, 53310, 57200, 56000, 63316, 69741]
companyB = [37596, 42000, 45510, 47310, 51200, 55000, 60316, 65741]
companyC = [36596, 32010, 40510, 30310, 42200, 50100, 45116, 34741]
# vẽ trên 2 thằng trên 1 hệ trục tọa độ
plt.plot(years, companyA, color = 'green', linestyle = '--', marker='.', markersize=20)
plt.plot(years, companyB, color = 'red', linestyle = '-', marker='.', markersize=10)
plt.plot(years, companyC, color = 'blue')
plt.xlabel("Year")
plt.ylabel("Profit(USD)")
# chú thích màu và kiểu
plt.legend(["CompanyA","CompanyB","CompanyC"])
plt.show()

```

```python
def draw_multiline(): 
    plt.style.use('default')
    years = [2013, 2014, 2015, 2016, 2017, 2018, 2019, 2020]
    #profit of CompanyA
    companyA = [41596, 46000, 48510, 53310, 57200, 56000, 63316, 69741]
    companyB = [37596, 42000, 45510, 47310, 51200, 55000, 60316, 65741]
    companyC = [36596, 32010, 40510, 30310, 42200, 50100, 45116, 34741]
    # vẽ trên 2 thằng trên 1 hệ trục tọa độ
    plt.plot(years, companyA, color = 'green', linestyle = '--', marker='.', markersize=20)
    plt.plot(years, companyB, color = 'red', linestyle = '-', marker='.', markersize=10)
    # bieu thi dang cot
    plt.bar(years, companyC, color = 'blue')
    plt.xlabel("Year")
    plt.ylabel("Profit(USD)")
    # chú thích màu và kiểu
    plt.legend(["CompanyA","CompanyB","CompanyC"])
    # đồ thị chia lưới và phân thành lưới nhỏ
    plt.grid(True)
    # lưu đồ thị thành file ảnh
    plt.savefig("multiline.png")
    plt.show()

draw_multiline()

```
