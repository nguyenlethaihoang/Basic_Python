```python
# import library
import numpy as np
import matplotlib.pyplot as plt
```

```python
# height (cm)
X = np.array([[147, 150, 153, 158, 163, 165, 168, 170, 173, 175, 178, 180, 183]]).T
# weight (kg)
Y = np.array([[49, 50, 51,  54, 58, 59, 60, 62, 63, 64, 66, 67, 68]]).T
# visualize data
plt.plot(X, Y, 'ro',)
plt.axis([140, 200, 40, 80])
plt.xlabel("Height (cm)")
plt.ylabel("Weight (kg)")
plt.show()
```

```python
# weight = w_1 * height + w_0 

## 1. BUIDING X BAR

# create array with value, X.shape[0] : number of dimension in the array
one = np.ones((X.shape[0], 1))     
# Join a sequence of array along axis(0,1)
Xbar = np.concatenate((one, X), axis = 1)


## 2. CALCULATING WEIGHTS OF THE FITTING LINE

# nhân ma trận Xbar chuyển vị và Xbar
A = np.dot(Xbar.T, Xbar)
b = np.dot(Xbar.T, Y)
w = np.dot(np.linalg.pinv(A),b)
print('w = ', w)

## 3. PREPARING THE FITTING LINE 
w_0 = w[0][0]
w_1 = w[1][0]
x0 = np.linspace(145, 185, 2)
y0 = w_0 + w_1*x0

## 4. DRAWING THE FITTING LINE
plt.plot(X.T, Y.T, 'ro') #data
plt.plot(x0, y0) # the fitting line
plt.axis([140, 190, 45, 75])
plt.xlabel("Height (cm)")
plt.ylabel("Weight (kg)")
plt.show()
```

```python
## PREDICT

y1 = w_1*155 + w_0
y2 = w_1*160 + w_0

print( u'Predict weight of person with height 155 cm: %.2f (kg), real number: 52 (kg)'  %(y1) )
print( u'Predict weight of person with height 160 cm: %.2f (kg), real number: 56 (kg)'  %(y2) )
```

```python
from sklearn import datasets, linear_model

## Fit the model by Linear Regression

# fit_intercept = False for calculating the bias
regr = linear_model.LinearRegression(fit_intercept=False) 
regr.fit(Xbar, Y)

# Compare two results
print( 'Solution found by scikit-learn  : ', regr.coef_ )
print( 'Solution found by (5): ', w.T)

```
